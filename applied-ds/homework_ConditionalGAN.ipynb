# Данные

Будем использовать данные телескопа MAGIC из UCI репозитория https://archive.ics.uci.edu/ml/datasets/MAGIC+Gamma+Telescope. Каждый объект в данных - параметры одного изображения кластера и метка этого кластера (фотон или адрон):


0. Length: major axis of ellipse [mm]
1. Width: minor axis of ellipse [mm]
2. Size: 10-log of sum of content of all pixels [in #phot]
3. Conc: ratio of sum of two highest pixels over fSize [ratio]
4. Conc1: ratio of highest pixel over fSize [ratio]
5. Asym: distance from highest pixel to center, projected onto major axis [mm]
6. M3Long: 3rd root of third moment along major axis [mm]
7. M3Trans: 3rd root of third moment along minor axis [mm]
8. Alpha: angle of major axis with vector to origin [deg]
9. Dist: distance from origin to center of ellipse [mm]
10. class: g,h # gamma (signal), hadron (background)

# Conditional WGAN

Мы будем использовать `Conditional WGAN`, который изображен на рисунке. В качестве условия `y` мы будем использовать метку класса: **0** - адрон, **1** - фотон. Таким образом, мы будем сообщать генератору для какой частицы нужно генерировать параметры изображения.

<center><img src="https://github.com/hse-ds/iad-applied-ds/blob/master/2023/hw/hw1/img/cgan.png?raw=1" width="800"></center>

Генератор $\hat{x} = G(z, y)$ будет принимать на вход шумовой вектор $z$ и вектор условий $y$, а выдавать будет сгенерированный (фейковый) вектор параметров $\hat{x}$.

Дискриминатор $D(x, y)$ будет принимать на вход вектор параметров $x$ и вектор условий $y$, а возвращать будет рациональное число.

Обучать `Conditional WGAN` будем с такой функцией потерь:

$$L(G, D) = -\frac{1}{n} \sum_{x_i \in X, y_i \in y} D(x_i, y_i) + \frac{1}{n} \sum_{z_i \in Z, y_i \in y} D(G(z_i, y_i), y_i) \to \max_G \min_D$$

```python
import torch
import torch.nn as nn
import torch.nn.functional as F
from torch.autograd import Variable
from torch.utils.data import DataLoader, TensorDataset

DEVICE = torch.device("cuda" if torch.cuda.is_available() else "cpu")
```

- ReLU функцию активации;
- Выходной слой.

```python
class Generator(nn.Module):
    def __init__(self, n_inputs, n_outputs):
        super(Generator, self).__init__()

        self.model = nn.Sequential(
            nn.Linear(n_inputs, 100),
            nn.BatchNorm1d(100),
            nn.LeakyReLU(0.2),
            nn.Linear(100, 100),
            nn.BatchNorm1d(100),
            nn.LeakyReLU(0.2),
            nn.Linear(100, n_outputs),
            nn.Tanh()
        )

    def forward(self, z, y):
        zy = torch.cat((z, y), dim=1)
        return self.model(zy)
```

## Задание 4 (0.5 балл)

Реализуйте функцию для генерации новый объектов $X$ по вектору условий $y$.

```python
def generate(generator, y, latent_dim):
    y = torch.tensor(y, dtype=torch.float, device=DEVICE)
    z = torch.normal(0, 1, (y.shape[0], latent_dim)).to(DEVICE)

    with torch.no_grad():
      X_fake = generator(z, y)

    return X_fake.cpu().detach().numpy()
```

```python
class Discriminator(nn.Module):
    def __init__(self, n_inputs):
        super(Discriminator, self).__init__()

        self.model = nn.Sequential(
            nn.Linear(n_inputs, 100),
            nn.LeakyReLU(0.2),
            nn.Linear(100, 100),
            nn.Dropout(0.4),
            nn.LeakyReLU(0.2),
            nn.Linear(100, 1),
            nn.Sigmoid()
        )

    def forward(self, x, y):
        xy = torch.cat((x, y), dim=1)
        return self.model(xy)
```

```python
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.linear_model import LinearRegression
from sklearn.metrics import accuracy_score
```

```python
def other_agent_score(X_real, y_real, X_fake, y_fake):
    X = np.concatenate((X_real, X_fake), axis=0)
    y = np.concatenate((y_real, y_fake), axis=0)

    X_train, X_test, y_train, y_test = train_test_split(
        X, y, stratify=y, test_size=0.25, shuffle=True, random_state=11
    )

    clf_lin = LinearRegression()
    clf_boost = GradientBoostingClassifier()

    clf_lin.fit(X_train, y_train)
    clf_boost.fit(X_train, y_train)

    lin_reg_score = accuracy_score(y_test, clf_lin.predict(X_test) > 0.5)
    boosting_score = accuracy_score(y_test, clf_boost.predict(X_test))

    print('Linear regression score: ' + str(lin_reg_score))
    print('Boosting score: ' + str(boosting_score))
```

## Задание 7 (1.5 балла)

Реализуйте класс для обучения генеративной модели.

- Подсказка 1: не забывайте ограничивать веса дискриминатора. Для этого используйте `p.data.clamp_(-0.01, 0.01)`, где `p` веса дискриминатора.
- Подсказка 2: `n_critic` - число итераций обучения дискриминатора на одну итерацию обучения генератора.
- Подсказка 3: Используйте `X_tensor = torch.tensor(X_numpy, dtype=torch.float, device=DEVICE)` для перевода numpy в тензор.

```python
from IPython.display import clear_output
```

```python
class Fitter(object):
    def __init__(
        self,
        generator,
        discriminator,
        batch_size = 32,
        n_epochs = 10,
        latent_dim = 10,
        n_critic = 5,
        graph_iter = 10,
    ):

        self.generator = generator
        self.discriminator = discriminator
        self.batch_size = batch_size
        self.n_epochs = n_epochs
        self.latent_dim = latent_dim
        self.n_critic = n_critic
        self.graph_iter = graph_iter

        self.opt_gen = torch.optim.RMSprop(self.generator.parameters(), lr=0.0001)
        self.opt_disc = torch.optim.RMSprop(self.discriminator.parameters(), lr=0.0001)

        self.generator.to(DEVICE)
        self.discriminator.to(DEVICE)

    def fit(self, X, y):

        # numpy to tensor
        X_real = torch.tensor(X, dtype=torch.float, device=DEVICE)
        y_cond = torch.tensor(y, dtype=torch.float, device=DEVICE)

        # tensor to dataset
        dataset_real = TensorDataset(X_real, y_cond)

        self.loss_history = []

        gen_loss_trace = []
        disc_loss_trace = []

        self.generator.train(True)
        self.discriminator.train(True)

        iter_i = 0

        for epoch in range(self.n_epochs):

            for i, (real_batch, cond_batch) in enumerate(
                DataLoader(dataset_real, batch_size=self.batch_size, shuffle=True)
                ):
                real_batch.to(DEVICE)
                cond_batch.to(DEVICE)

                num_objects = cond_batch.shape[0]

                for n in range(self.n_critic):
                  self.opt_disc.zero_grad()

                  z = torch.normal(0, 1, (num_objects, self.latent_dim)).to(DEVICE)
                  gen_objects = self.generator(z, cond_batch)

                  disc_loss = -torch.mean(self.discriminator(real_batch, cond_batch)) + torch.mean(self.discriminator(gen_objects, cond_batch))

                  disc_loss.backward()
                  self.opt_disc.step()

                  if n == self.n_critic - 1: disc_loss_trace.append((iter_i, disc_loss.item()))

                  for p in self.discriminator.parameters():
                    p.data.clamp_(-0.01, 0.01)

                self.opt_gen.zero_grad()

                z = torch.normal(0, 1, (num_objects, self.latent_dim)).to(DEVICE)
                gen_objects = self.generator(z, cond_batch)

                gen_loss = -torch.mean(self.discriminator(gen_objects, cond_batch))

                gen_loss.backward()
                self.opt_gen.step()

                gen_loss_trace.append((iter_i, gen_loss.item()))

                iter_i += 1

                # show dynamic graphs
                if iter_i % self.graph_iter == 0:
                    clear_output(wait=True)
                    plt.figure(figsize=(10, 5))

                    plt.subplot(1, 2, 1)
                    plt.xlabel("Iteration")
                    plt.ylabel("Generator loss")
                    plt.plot([p[0] for p in gen_loss_trace], [p[1] for p in gen_loss_trace], alpha=0.5, color='y')
                    plt.grid(linestyle="--")

                    plt.subplot(1, 2, 2)
                    plt.xlabel("Iteration")
                    plt.ylabel("Discriminator loss")
                    plt.plot([p[0] for p in disc_loss_trace], [p[1] for p in disc_loss_trace], alpha=0.5, color='g')
                    plt.grid(linestyle="--")

                    plt.show()

            # caiculate and store loss after an epoch
            Z_noise = torch.normal(0, 1, (len(X_real), self.latent_dim)).to(DEVICE)

            with torch.no_grad():
              X_fake = self.generator(Z_noise, y_cond)

            loss_epoch = torch.mean(self.discriminator(X_real, y_cond)).mean() - torch.mean(
                self.discriminator(X_fake, y_cond)
            )

            self.loss_history.append(loss_epoch.detach().cpu())

            # other agent score

            rand_idx = torch.randint(0, X_real.shape[0], (1000,))
            other_agent_score(X_real[rand_idx, :].cpu().detach().numpy(),
                              np.array([1] * 1000),
                              X_fake[rand_idx, :].cpu().detach().numpy(),
                              np.array([0] * 1000)
            )

        self.generator.train(False)
        self.discriminator.train(False)
```

## Обучение
Обучим модель на данных.

```python
%%time
latent_dim = 10
generator = Generator(n_inputs=latent_dim + y.shape[1], n_outputs=X_train.shape[1])
discriminator = Discriminator(n_inputs=X_train.shape[1] + y.shape[1])

fitter = Fitter(
    generator,
    discriminator,
    batch_size=64,
    n_epochs=30,
    latent_dim=latent_dim,
    n_critic=5,
    graph_iter=100
)
fitter.fit(X_train, y_train)
```

```python
class DiffusionGenerator(nn.Module):
    def __init__(self, n_inputs, n_outputs):
        super(DiffusionGenerator, self).__init__()

        self.model = nn.Sequential(
            nn.Linear(n_inputs, 100),
            nn.BatchNorm1d(100),
            nn.ReLU(),
            nn.Linear(100, 100),
            nn.BatchNorm1d(100),
            nn.ReLU(),
            nn.Linear(100, n_outputs),
            nn.Tanh()
        )

    def forward(self, z, y):
        zy = torch.cat((z, y), dim=1)
        return self.model(zy)
```

## Задание 12 (0.5 балла)

Напишите функцию для генерации нового объекта с помощью обученной модели.

```python
def generate_with_diffusion(model, y, latent_dim, scheduler):
    y = torch.tensor(y, dtype=torch.float, device=DEVICE)
    X_fake = torch.randn(y.shape[0], latent_dim).to(DEVICE)

    for i, t in enumerate(scheduler.timesteps):
      with torch.no_grad():
        pred = model(X_fake, y)
      X_fake = scheduler.step(pred, t, X_fake).prev_sample

    return X_fake.cpu().detach().numpy()
```

## Задание 13 (2 балла)

Напишите класс обучения диффузионной модели и обучите модель, после чего опишите полученные результаты. В качестве подсказки - опирайтесь на семинар по диффузионным моделям. Вы можете изменять некоторые части кода для вашего удобства, но оставляйте в таком случаи комментарии, пожалуйста.

```python
class DiffusionFitter(object):
    def __init__(
        self,
        model,
        scheduler,
        batch_size = 32,
        n_epochs = 10,
        latent_dim = 1,
        lr = 0.0001,
    ):

        self.model = model
        self.scheduler = scheduler
        self.batch_size = batch_size
        self.n_epochs = n_epochs
        self.latent_dim = latent_dim
        self.lr = lr

        self.opt_gen = torch.optim.RMSprop(self.model.parameters(), lr=self.lr)

        self.model.to(DEVICE)

    def fit(self, X, y):

        # numpy to tensor
        X_real = torch.tensor(X, dtype=torch.float, device=DEVICE)
        y_cond = torch.tensor(y, dtype=torch.float, device=DEVICE)

        # tensor to dataset
        dataset_real = TensorDataset(X_real, y_cond)

        self.model.train(True)

        self.loss_history = []

        for epoch in range(self.n_epochs):
            loss_epoch = []
            for i, (real_batch, cond_batch) in enumerate(
                DataLoader(dataset_real, batch_size=self.batch_size, shuffle=True)
            ):

                real_batch = real_batch.to(DEVICE)
                cond_batch = cond_batch.to(DEVICE)

                noise = torch.randn_like(real_batch)
                timesteps = torch.randint(0, 999, (real_batch.shape[0],)).long().to(DEVICE)
                noisy_batch = self.scheduler.add_noise(real_batch, noise, timesteps)

                pred = self.model(noisy_batch, cond_batch)

                loss = nn.MSELoss()(pred, noise)

                self.opt_gen.zero_grad()
                loss.backward()
                self.opt_gen.step()

                loss_epoch.append(loss.item())

            # calculate and store loss after an epoch

            self.loss_history.append(sum(loss_epoch) / len(loss_epoch))
            loss_epoch.clear()

            # other agent score

            clear_output(wait=True)

            rand_idx = torch.randint(0, X_real.shape[0], (1000,))
            X_fake = generate_with_diffusion(self.model, y_train, latent_dim, self.scheduler)

            other_agent_score(X_real[rand_idx, :].cpu().detach().numpy(),
                              np.array([1] * 1000),
                              X_fake[rand_idx, :],
                              np.array([0] * 1000)
            )

        self.model.train(False)
```

```python
%%time
latent_dim = 10
model = DiffusionGenerator(n_inputs=latent_dim + y.shape[1], n_outputs=X_train.shape[1])

diffusionFitter = DiffusionFitter(
    model,
    noise_scheduler,
    batch_size=64,
    n_epochs=100,
    latent_dim=latent_dim,
    lr=0.0001,
)
diffusionFitter.fit(X_train, y_train)
```
